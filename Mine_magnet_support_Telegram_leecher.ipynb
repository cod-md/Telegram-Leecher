```python

# @title <font color=o> ‚è¨ Download to colab

!apt-get install aria2

import requests

import os

import re

import ipywidgets as widgets

from IPython.display import display, clear_output, HTML

import subprocess

def download_google_drive_link(url, save_directory):

    try:

        # Extract the file ID from the Google Drive link

        file_id = url.split('/')[-2]

        # Construct the direct download link

        direct_download_url = f"https://drive.google.com/uc?id={file_id}"

        response = requests.get(direct_download_url, stream=True)

        if response.status_code == 200:

            # Get the original filename from the response headers

            content_disposition = response.headers.get("content-disposition")

            if content_disposition:

                # Use regular expression to extract the filename

                filename_match = re.search(r'filename\*?=["]?([^"]*)["]?(; filename\*=UTF-8\'\'(.*))?', content_disposition)

                if filename_match:

                    filename = filename_match.group(1) or filename_match.group(3)

                    filename = filename.strip('"')

                else:

                    filename = file_id

            else:

                filename = file_id

            filename = os.path.join(save_directory, filename)

            with open(filename, 'wb') as f:

                for chunk in response.iter_content(chunk_size=1024):

                    if chunk:

                        f.write(chunk)

            clear_output()

            display(HTML(f"<b>Downloaded:</b> {url} => {filename}"))

        else:

            clear_output()

            display(HTML(f"<b>Failed to download:</b> {url} (Status code: {response.status_code})"))

    except Exception as e:

        clear_output()

        display(HTML(f"<b>Error downloading {url}:</b> {str(e)}"))

# Create widgets for entering multiple URLs, custom names, a single custom extension, and the download button

url_input = widgets.Textarea(placeholder='Enter URLs or magnet links (one per line)')

name_input = widgets.Textarea(placeholder='Enter custom names (one per line)')

extension_input = widgets.Text(placeholder='Enter custom extension (e.g., mp4)')

download_button = widgets.Button(description='Download')

# Define the custom extension input widget

extension_input = widgets.Text(placeholder='Enter custom extension (e.g., mp4)')

# Define the download function for the button click

def on_download_button_click(b):

    urls = url_input.value.split('\n')

    names = name_input.value.split('\n')

    custom_extension = extension_input.value.strip()

    for url, custom_name in zip(urls, names):

        url = url.strip()  # Remove leading/trailing whitespace

        custom_name = custom_name.strip()  # Remove leading/trailing whitespace

        if url:

            if "drive.google.com" in url:

                download_google_drive_link(url, save_directory.value)

            else:

                download_video(url, save_directory.value, custom_name, custom_extension)

download_button.on_click(on_download_button_click)

# Create a widget for specifying the save directory with the default value

save_directory = widgets.Text(placeholder='Save directory', value='/content/')

# Display widgets

display(url_input, name_input, extension_input, download_button, save_directory)
```
